# Salary Analysis

```python
sns.histplot(data=df, x='MonthlyIncome',bins=10,kde=True)
plt.title('Monthly Income')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df,x='JobRole',y='MonthlyIncome')
plt.title('Salaries for each job role')
plt.xticks(rotation = 90)
plt.show()
```

```python
#Satisfaction Rates
sns.histplot(data=df, x='JobSatisfaction',bins=10,kde=True)
plt.title('Satisfaction rates for jobs')
plt.show()
```

```python
sns.histplot(data=df, x='PerformanceRating',bins=10,kde=True)
plt.title('Performance Rating')
plt.show()
```

```python
sns.histplot(data=df, x='YearsAtCompany',bins=10,kde=True)
plt.title('Employee life time at company')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df,x='Department',y='Age')
plt.title('Ages for each Department')
plt.xticks(rotation =90)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df,x='JobRole',hue='Attrition')
plt.title('Attrition for each job')
plt.xticks(rotation =90)
plt.show()
```
